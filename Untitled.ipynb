```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import joblib

# --- 1. Завантажуємо дані, зібрані на попередньому етапі ---
try:
    df = pd.read_csv("data/EURUSD_15m_history.csv", parse_dates=['ts'])
    print(f"Дані успішно завантажено. Кількість рядків: {len(df)}")
except FileNotFoundError:
    print("Помилка: файл 'data/EURUSD_15m_history.csv' не знайдено. Переконайтесь, що він знаходиться в папці 'data'.")
    # Зупиняємо виконання, якщо файлу немає
    exit()

# --- 2. Обираємо характеристики (features) для кластеризації ---
# Ми використовуємо волатильність (ATR), силу тренду (ADX) та моментум (RSI)
features = df[['ATR', 'ADX', 'RSI']].copy()
print("Характеристики для аналізу обрано.")

# --- 3. Масштабування даних ---
# Це важливий крок, щоб індикатори з різними діапазонами (0-100 і 0-0.001) мали однакову вагу
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
print("Дані масштабовано.")

# --- 4. Тренування моделі K-Means ---
# Ми шукаємо 4 основні режими ринку
K = 4
kmeans = KMeans(n_clusters=K, random_state=42, n_init='auto')
df['market_regime'] = kmeans.fit_predict(scaled_features)
print(f"Модель K-Means натренована, знайдено {K} кластерів (режимів ринку).")

# --- 5. Аналіз та опис знайдених режимів ---
print("\n--- Характеристики знайдених режимів ринку ---")
# centroids_df створюється для зручності аналізу
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
centroids_df = pd.DataFrame(centroids, columns=features.columns)
centroids_df['cluster_size'] = df['market_regime'].value_counts().sort_index()

# Даємо назви кластерам на основі їх характеристик
regime_names = {}
for i, row in centroids_df.iterrows():
    if row['ADX'] < 20:
        regime_names[i] = "Млявий флет (низький тренд)"
    elif row['ATR'] > centroids_df['ATR'].quantile(0.75):
        regime_names[i] = "Шторм (висока волатильність)"
    elif row['RSI'] > 55:
        regime_names[i] = "Бичачий тренд"
    elif row['RSI'] < 45:
        regime_names[i] = "Ведмежий тренд"
    else:
        regime_names[i] = f"Змішаний режим {i}"

centroids_df['regime_name'] = centroids_df.index.map(regime_names)
print(centroids_df[['regime_name', 'ATR', 'ADX', 'RSI', 'cluster_size']])

# --- 6. Зберігаємо навчену модель і скейлер ---
joblib.dump(kmeans, 'market_regime_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("\nНавчена модель та скейлер збережені у файли 'market_regime_model.pkl' та 'scaler.pkl'.")
```
